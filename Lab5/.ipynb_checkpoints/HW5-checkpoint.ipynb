```python
label1_mean = np.mean(label1, axis=0)
label2_mean = np.mean(label2, axis=0)
std1 = (label1 - label1_mean)
std2 = (label2 - label2_mean)
S1 = np.matmul(std1.T, std1)
S2 = np.matmul(std2.T, std2)
SW = S1 + S2
SW_inv = np.linalg.inv(SW)
mean_diff = label1_mean - label2_mean
w = np.matmul(SW_inv, mean_diff.T)
label1_proj = np.matmul(label1, w)
label2_proj = np.matmul(label2, w)
```

```python
label1_y = np.array([1] * len(label1))
label2_y = np.array([2] * len(label2))
y = np.concatenate((label1_y, label2_y))
```

```python
labels = np.concatenate((label1, label2))
clf = LinearDiscriminantAnalysis()
clf.fit(labels, y)
label1_sk = clf.transform(label1)
label2_sk = clf.transform(label2)
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
ax.scatter(label1_proj, np.ones(len(label1_proj)), c='r', marker='o', label="label1")
ax.scatter(label2_proj, np.ones(len(label2_proj)), c='b', marker='x', label="label2")
ax.legend()
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)

ax.scatter(label1_sk, np.ones(len(label1_sk)), c='r', marker='o', label="label1")
ax.scatter(label2_sk, np.ones(len(label2_sk)), c='b', marker='x', label="label2")
ax.legend()
plt.show()
```

## Part B)

```python
from sklearn.linear_model import LogisticRegression
log_model = LogisticRegression()
X = weekly.loc[:,'Lag1':'Volume']
y = weekly.loc[:,'Direction']
log_model.fit(X, y)
log_model.coef_
```

Based on the coefficients generated and these predictor values having relatively similar ranges, none of the variables seem statistically significant.

## Part C)

```python
from sklearn.metrics import confusion_matrix
pred = log_model.predict(X)
conf_mat = confusion_matrix(y, pred)
conf_mat
```

The confusion matrix is saying that for every time we predicted a certain class we got a certain amount right and certain amount wrong. This is a binary classifier so it corresponds that the matrix is 2x2 because for every class it can be predicted correctly or incorrectly.

Specifically, the matrix is saying for every time the classifier predicted Down it was right 55 times and wrong 47 times. For every time the classifier predicted Up it was right 558 times and wrong 429 times. So the classifier is doing slightly better than 50% accuracy. It also shows that Up is predicted way more times than Down even when it's wrong.

## Part D)

```python
X_train = weekly.loc[weekly['Year'] <= 2008].loc[:,'Lag2'].reshape(1,-1).T
X_test = weekly.loc[weekly['Year'] > 2008].loc[:,'Lag2'].reshape(1,-1).T
y_train = weekly.loc[weekly['Year'] <= 2008].loc[:,'Direction']
y_test = weekly.loc[weekly['Year'] > 2008].loc[:,'Direction']
```

```python
lag2_model = LogisticRegression()
lag2_model.fit(X_train, y_train)
pred_log = lag2_model.predict(X_test)
conf_mat2 = confusion_matrix(y_test, pred_log)
frac = float(conf_mat2[0,0] + conf_mat2[1,1]) / len(y_test)
print("Fraction correct = " + str(frac))
conf_mat2
```

## Part E)

```python
lda = LinearDiscriminantAnalysis()
lda.fit(X_train, y_train)
pred_lda = lda.predict(X_test)
conf_lda = confusion_matrix(y_test, pred_lda)
frac = float(conf_lda[0,0] + conf_lda[1,1]) / len(y_test)
print("Fraction correct = " + str(frac))
conf_lda
```

## Part F)

```python
from sklearn.discriminant_analysis import QuadraticDiscriminantAnalysis as QDA
qda = QDA()
qda.fit(X_train,y_train)
pred_qda = qda.predict(X_test)
conf_qda = confusion_matrix(y_test, pred_qda)
frac = float(conf_qda[0,0] + conf_qda[1,1]) / len(y_test)
print("Fraction correct = " + str(frac))
conf_qda
```

## Part G)

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=1)
knn.fit(X_train, y_train)
pred_knn = knn.predict(X_test)
conf_knn = confusion_matrix(y_test, pred_knn)
frac = float(conf_knn[0,0] + conf_knn[1,1]) / len(y_test)
print("Fraction correct = " + str(frac))
conf_knn
```
